```python
import re
import matplotlib.pyplot as plt
from collections import Counter

# Example text data representing the 11 volumes
volumes_data = {
    'Volume 1': 'This is a simple example for volume one. Example text for volume one.',
    'Volume 2': 'Here we have more example text for volume two.',
    'Volume 3': 'This is volume three with different text.',
    'Volume 4': 'Volume four has its own unique set of words.',
    'Volume 5': 'Now we are looking at volume five with other examples.',
    'Volume 6': 'Text data for volume six here, continuing the story.',
    'Volume 7': 'Volume seven introduces new concepts and words.',
    'Volume 8': 'In volume eight, we discuss more examples and details.',
    'Volume 9': 'Volume nine brings more ideas to the table.',
    'Volume 10': 'This is volume ten with its own set of information.',
    'Volume 11': 'Finally, volume eleven concludes with final examples.',
}

# Simple text preprocessing function using regular expressions
def preprocess_text(text):
    # Convert text to lowercase and tokenize using regular expression (split by non-alphabetical characters)
    text = text.lower()
    tokens = re.findall(r'\b\w+\b', text)  # This keeps only words consisting of alphabets
    return tokens

# Function to analyze word frequencies in volumes
def analyze_word_frequencies(volumes_data):
    word_frequencies = {}
    for volume_name, text in volumes_data.items():
        tokens = preprocess_text(text)
        word_count = Counter(tokens)
        word_frequencies[volume_name] = word_count
    return word_frequencies

# Analyze word frequencies
word_frequencies = analyze_word_frequencies(volumes_data)

# Plotting the word frequencies for the top words in all volumes
top_words = []
for volume_name, word_count in word_frequencies.items():
    top_words.append(dict(word_count.most_common(10)))  # Top 10 words for each volume

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
for i, top_word in enumerate(top_words):
    words = list(top_word.keys())
    counts = list(top_word.values())
    ax.barh(words, counts, label=f'Volume {i+1}')

ax.set_xlabel('Word Count')
ax.set_title('Top 10 Words Frequency Across Volumes')
ax.legend()
plt.show()
```
